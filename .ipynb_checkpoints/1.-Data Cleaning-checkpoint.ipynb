## Context

In this lab we are going to work with data to understand the characteristics of a diamond that are most likely to influence its price. In this first part of the lab, we will explore and clean our data. 

The dataset we will be using is comprised of approximately 54k rows and 11 different columns. As always, a row represents a single observation (in this case a diamond) and each of the columns represent a different feature of a diamond.

The following codebook was provided together with the dataset to clarify what each column represents:


| Column  | Description  |
|---|---|
| Price  | Price in US dollars (326-18,823)  |
| Carat  | Weight of the diamond (0.2--5.01)  |
| Cut  | Quality of the cut (Fair, Good, Very Good, Premium, Ideal)  |
| Color  | Diamond colour, from J (worst) to D (best)  |
| Clarity  | A measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))   |
| x  | Length in mm (0--10.74)  |
| y  | Width in mm (0--58.9)  |
| z  | Depth in mm (0--31.8)  |
| Depth  | Total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)  |
| Table  | Width of top of diamond relative to widest point (43--95)  |

You have probably noticed that the columns x, y and z have a minimum value of 0. This means that there are one or more rows (or observations) in our dataset that are supposedly representing a diamond that has lenght, width or depth of 0. Considering that we're talking about a physical object, this is impossible!

Now let's proceed to check the rows that have a value of 0 in any of the x, y or z columns. By doing this we want to check if the data we are missing can be obtained using the data that we do have.

**Check the columns with `x`, `y` and `z` with value 0 in all of them and comment what you see**

```python
#your code here
diamonds.loc[(diamonds['x'] == 0) & (diamonds['y'] == 0) & (diamonds['z'] == 0)]

# There are 7 row where all three measures are missing.
```

```python
diamonds.loc[(diamonds['x'] == 0) | (diamonds['y'] == 0)| (diamonds['z'] == 0)]

# There are 20 rows where eigher of three measures or all of them are missing. Majority of them have x & y values but z is missing.
```

As you can see, we have 20 rows that have a value of 0 in some or all the aforementioned columns.
Most of them (12) are missing the z value, which we can obtain using the columns depth, x and y. 

20 rows with issues represent just 0.03% of our data (20 out of 53940) so it wouldn't be a big deal to remove them. Still, lets try to keep all the data we have. 

For those 12 rows, we will create a function that applies the formula given in the codebook and get the value of z. We will drop the other rows (8), since they are missing all 3 values or 2 of them.

**Create a function named `calculate_z` that applies the function in the codebook to one single row you give to the function**

```python
#your code here

# Total depth percentage = z / mean(x, y) = 2 * z / (x + y)
# hense z = (d*(x+y))/2
def calculate_z(row):
    return round((row['depth']/100*(row['x']+row['y']))/2, 2)
    
# calculate_z(diamonds.iloc[2207])
```

**Apply it just to the rows with incorrect values**

```python
diamonds['z'] = np.where(diamonds['z'] == 0, diamonds.apply(lambda row: calculate_z(row), axis=1), diamonds['z'])
```

```python
# Check one known row is updated.
diamonds.iloc[2207]
```

```python
# Summary check to confirm there are no zeros in z anymore.
diamonds.loc[(diamonds['x'] == 0) | (diamonds['y'] == 0)| (diamonds['z'] == 0)]
# There are still 8 rows with missing values in x or/and y so z can't be calculated as well.
```

If we leave the other 8 values as they are, it would negatively affect our analysis, because these are data that do not make logical sense. Therefore it is better to consider those values as NaN values, since they are probably the result of a mistake or error during process of measuring and storing these values in a dataset.

To replace them we can use the pandas .replace() method and np.NaN.

**Replace the zero values in the `z` column for a NaN**

```python
#your code here
# It actually needs to be done for x, y & z and not only for z if we want to be able to see correct min values before dropping
# any rows as it's requested in section below.

# d['z'].replace(0, np.NaN) # Why doesn't work?
diamonds['z'] = np.where(diamonds['z'] == 0, np.NaN, diamonds['z'])
diamonds['x'] = np.where(diamonds['x'] == 0, np.NaN, diamonds['x'])
diamonds['y'] = np.where(diamonds['y'] == 0, np.NaN, diamonds['y'])

# Check known row that it worked.
diamonds.iloc[11963]
```

# Bonus: check the new z values
Since we need to be 100% sure of our data, let's create a function that validates our z. To do so, we will use the same formula, but this time we will calculate the value of depth with the new value assigned to z.

**Create a function named `validate_z` that compares the `z`  in cells above with the one thrown by the formula and run it with the rows you changed in the cells above**

```python
#your code here
# Total depth percentage = z / mean(x, y) = 2 * z / (x + y)
def validate_z(row):
    d_calc = round(2*row['z']/(row['x']+row['y'])*100, 1)    # Calculate depth % based on x, y, z
    return round((d_calc - row['depth']), 1)                 # Compare calculated value with given value
    
# validate_z(diamonds.iloc[0])
```

```python
# I would like to check all my depth values and not only in rows where I changed z:
diamonds['depth_variance'] = diamonds.apply(lambda row: validate_z(row), axis=1)
```

```python
diamonds.iloc[13601]
```

```python
# Let's say variance below 1% is irrelevant and how many relevant variances we have.
check = diamonds.loc[(diamonds['depth_variance'] > 1) | (diamonds['depth_variance'] < -1)]
check
# It's actually quite a lot, 72 values or 0.13% of the population.
```

```python
check.pivot_table(index=['cut', 'color', 'clarity'], values=['depth_variance'])
# For some categories average variance is above 10%.
# Assuming that x, y & z are correct values I suggest to add new column 'depth_calculated' and compare main statistical measurements.
```

```python
def calculate_depth(row):
    return round(2*row['z']/(row['x']+row['y'])*100, 1)
```

```python
diamonds['depth_calc'] = diamonds.apply(lambda row: calculate_depth(row), axis=1)
diamonds.head()
```

```python
diamonds.describe()
```

```python
'''
As per summary above it seems it's not necessary to rewrite depth as major changes are in min and max and they don't seem to
be correct.
I will drop two columns I've added eearlier for check as they take space and in real world with real data it might matter.
Also it can be confusing for other users of data.
'''
```

```python
diamonds.drop(['depth_variance', 'depth_calc'], axis=1, inplace=True)
diamonds.head()
```

**Create a function named `calculate_y` to calculate `y` using `z` and `x` the same way you did above*
